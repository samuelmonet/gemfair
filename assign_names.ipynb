```python
import pandas as pd
```

```python
questions=pd.read_csv('questions.csv',sep='\t',index_col=0)
```

```python
questions=questions.T
questions
```

```python
questions.Idquest=questions.Idquest.astype('string')
questions.type=questions.type.astype('string')
questions.Treatment=questions.Treatment.astype('string')
questions.Idquest
```

```python
for i in range(len(kobo)):
    if kobo.loc[i]['label']==kobo.loc[i]['label']:
        for quest in questions.index:
            if kobo.loc[i]['label']==quest:
                questions.at[quest,'Idquest']=kobo.loc[i]['name']
                current=quest
                if kobo.loc[i]['type']=='decimal':
                    questions.at[quest,'type']='continue'
            elif kobo.loc[i]['label'] in quest and quest[2] not in '123456789':
                questions.at[quest,'Idquest']=current+' '+quest
                questions.at[quest,'Treatment']='cat'
            
```
